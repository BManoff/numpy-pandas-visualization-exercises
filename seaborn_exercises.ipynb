```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris
```

```python
#1.) What does the distribution of petal lengths look like?

sns.histplot(x = 'petal_length', hue = 'species', data = iris)
plt.title('Petal Length Distrobution')
plt.show()
```

```python
#2.) Is there a correlation between petal length and petal width?

#There is a positive correlation between length and width


sns.relplot(x = 'petal_length', y = 'petal_width', data = iris)
plt.title('Petal Length vs Petal Width')
plt.show()
```

```python
#3.) Would it be reasonable to predict species based on sepal width and sepal length? For this, 
# you'll visualize two numeric columns through the lense of a categorical column.

sns.barplot(x = 'petal_length', y = 'petal_width', data = iris)
plt.tight_layout
plt.title('Petal Length vs Petal Width')
plt.show()
```

```python
#4.) Which features would be best used to predict species?

sns.pairplot(iris, data = iris, hue = 'species')
plt.show()
```

```python
from env import host, user, password

def get_db_url(dbname, username = user, hostname = host, password = password): 
    url = f'mysql+pymysql://{user}:{password}@{host}/{dbname}'
    return url
```

```python

sql1 = '''
SELECT
    *
FROM orders
'''

chipotle_df = pd.read_sql(sql1, get_db_url('chipotle', user, host, password,))
chipotle_df

```

```python
chipotle_df['item_price'] = chipotle_df.item_price.str.replace('$', '', regex=False).astype('float')
```

```python
top_4 = chipotle_df.groupby('item_name').quantity.agg('sum').sort_values(ascending=False).head(4).reset_index()
top_4
```

```python
sns.barplot(x = 'item_name', y = 'quantity', data = top_4)

```

```python
revenue = chipotle_df[chipotle_df.item_name.isin(top_4.index)].groupby('item_name').item_price.sum()
revenue = revenue.sort_values(ascending = False).reset_index()
revenue
```

```python
sns.barplot(y = 'item_name', x = 'item_price', data = revenue)
plt.title('Most popular items and revenue')

plt.ylabel("") # to get remove the column_name
plt.xlabel("Gross Revenue")
plt.suptitle("Revenue from Top 4 selling Items")
```
