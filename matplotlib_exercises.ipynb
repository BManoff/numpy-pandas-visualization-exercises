```python
import matplotlib.pyplot as plt
import math
```

```python
x = range(0, 10)
y = [(x ** 2 - x + 2) for x in x]
plt.plot(x,y)

plt.xlabel('$X$')
plt.ylabel('$Y$')

plt.annotate("origin", xy = (0,0), arrowprops = {}, xytext=(0, 10))

#plt.savefig('fig1.png') <-- saves image in folder for later
```

```python
plt.subplot(221)
plt.title("$Y = x$")
x = range(0, 10)
y = [math.sqrt(x) for x in x]

plt.plot(x,y)

plt.xlabel('$X$')
plt.ylabel('$Y$')

plt.subplot(222)
plt.title("Y = X cubed")
x = range(0, 10)
y = [x ** 3 for x in x]

plt.plot(x,y)

plt.xlabel('$X$')
plt.ylabel('$Y$')

plt.subplot(223)
plt.title("Y = 2 to the X power")
x = range(0, 10)
y = [2 ** x for x in x]

plt.plot(x,y)

plt.xlabel('$X$')
plt.ylabel('$Y$')

plt.subplot(224)
plt.title("Y = 1 / (X + 1)")

x = range(0, 10)
y = [1/(x + 1) for x in x]

plt.plot(x,y)

plt.xlabel('$X$')
plt.ylabel('$Y$')

plt.tight_layout()
```

```python
x = list(range(0, 5))

y1 = [math.sqrt(x) for x in x]
y2 = [x ** 3 for x in x]
y3 = [2 ** x for x in x]
y4 = [1/(x + 1) for x in x]

plt.plot(x, y1, c='blue', label='Y = x')
plt.plot(x, y2, c='orange', label='Y = X cubed')
plt.plot(x, y3, c='green', label='Y = 2 to the X power')
plt.plot(x, y4, c='purple', label='Y = 1 / (X + 1)')

plt.legend(loc='upper left')
plt.title('Functions')
plt.xlabel('$X$')
plt.ylabel('$Y$')
```
