```python
import matplotlib.pyplot as plt
import math

with plt.xkcd():
    plt.figure(figsize = (10,5))
    plt.title('Big O Notation')
    plt.xlabel('$Elements$')
    plt.ylabel('$Operations$')



    x = range(1, 15)
    y1 = [1 for x in x]
    y2 = [math.log(x) for x in x]
    y3 = [x for x in x]
    y4 = [x * math.log(x) for x in x]
    y5 = [x ** 2 for x in x]
    y6 = [2 ** x for x in x]
    y7 = [math.factorial(x) for x in x]
    y8 = [x ** x for x in x]

    plt.plot(x, y1, c='blue', label='O(1)')
    plt.plot(x, y2, c='orange', label='O(log n)')
    plt.plot(x, y3, c='green', label='O(n)')
    plt.plot(x, y4, c='purple', label='O(n log n)')
    plt.plot(x, y5, c='black', label='O(n^2)')
    plt.plot(x, y6, c='red', label='O(2^n)')
    plt.plot(x, y7, c='yellow', label='O(n!)')
    plt.plot(x, y8, c='teal', label='O(n^n)')

    plt.legend(loc='upper left')
    plt.xlim(0, 10)
    plt.ylim(0, 50)
```
